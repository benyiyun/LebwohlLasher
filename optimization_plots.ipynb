```python
import numpy as np
import matplotlib.pyplot as plt

# Load the data
data = np.loadtxt("scaling_results.csv", delimiter=",", skiprows=1)
processes = data[:, 0]
runtimes = data[:, 1]

# Plot
plt.figure(figsize=(8, 5))
plt.plot(processes, runtimes, marker='o', linestyle='-', label="Runtime")

plt.xlabel("Number of Processes")
plt.ylabel("Runtime (seconds)")
plt.title("Runtime vs. Number of MPI Processes")
plt.xticks(processes.astype(int))
plt.grid()
plt.legend()

plt.savefig("scaling_plot.png")  # Save the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Load the data
cython_data = np.loadtxt("cython_scaling_results.csv", delimiter=",", skiprows=1)
lattice_sizes = cython_data[:, 0]
cython_runtimes = cython_data[:, 1]

mpi_data = np.loadtxt('mpi_scaling_results.csv', delimiter=",", skiprows=1)
mpi_runtimes = mpi_data[:, 1]

numba_data = np.loadtxt('numba_scaling_results.csv', delimiter=",", skiprows=1)
numba_runtimes = numba_data[:, 1]

vectorized_data = np.loadtxt('vectorized_scaling_results.csv', delimiter=",", skiprows=1)
vectorized_runtimes = vectorized_data[:, 1]

orig_data = np.loadtxt('orig_scaling_results.csv', delimiter= ",", skiprows=1)
orig_runtimes = orig_data[:, 1]

# Plot
plt.figure(figsize=(8, 5))
plt.plot(lattice_sizes, cython_runtimes, marker='o', linestyle='-', label="Cython Runtime", c='b')
plt.plot(lattice_sizes, mpi_runtimes, marker='o', linestyle='-', label="Mpi4py Runtime", c='g')
plt.plot(lattice_sizes, numba_runtimes, marker='o', linestyle='-', label="Numba Runtime", c='r')
plt.plot(lattice_sizes, vectorized_runtimes, marker='o', linestyle='-', label="Vectorized Runtime", c='c')
plt.plot(lattice_sizes, orig_runtimes, marker='o', linestyle='-', label="Original Runtime", c='k')

plt.xlabel("Lattice Size (N x N)")
plt.ylabel("Runtime (seconds)")
plt.title("Runtime vs. Lattice Size")
# plt.xscale("log")  # Use log scale if size varies significantly
# plt.yscale("log")  # Log-log scale helps see trends
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.legend()

plt.savefig("combined_scaling_plot.png")  # Save the plot
plt.show()
```
